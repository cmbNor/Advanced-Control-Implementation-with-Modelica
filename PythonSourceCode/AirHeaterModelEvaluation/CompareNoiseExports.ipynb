```python
#Import of libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Import the csv file that holds the exported data from simulating noise generator in modelica based on a External Object
dfEO = pd.read_csv('PIDairHeaterExternalObjectNoise.csv')
# Import the csv file that holds the exported data from simulating noise generator in modelica based on a Function
dfFunc = pd.read_csv('PIDairHeaterFunction.csv')

#Renaming columns name
dfEO.rename(columns = {'add.y':'EOOutput'}, inplace = True)
dfFunc.rename(columns = {'add.y':'FuncOutput'}, inplace = True)
```

```python
#Activate interactive shell
%config InteractiveShell.ast_node_interactivity = 'all'
```

```python
# Remove the percentiles from the output and display stats
perc = []
dfFunc.head()
dfFunc.describe(percentiles=perc)
dfEO.describe(percentiles=perc)
```

```python
# Plotting of the output of both simulations
fig, ax3 = plt.subplots(figsize=(12, 6));
plt.title('Difference in output noise signal', fontsize=20);

#Range of X-axises
ax3.set_ylim(19.4, 27);
ax3.set_xlim(0, 600);

# Set up the axis labels
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('Temperature [deg C]', color='k')

# Customize the major grid (outer grid)
ax3.grid(which='both', linestyle='-', linewidth='0.5', color='black')

# Set up data to be plotted with additional parameters
ax3.plot(dfFunc.time, dfFunc.FuncOutput, '-g', label='Function implementation', linewidth=1.5)
ax3.plot(dfEO.time, dfEO.EOOutput, '-r', label='External object implementation', linewidth=1.5)

# Configure placement of data legends
ax3.legend(loc='upper left')

# Save a copy of the plot
plt.savefig('pltCompareDiffNoiseImplementationsFull.png', dpi=300, bbox_inches='tight')

plt.show()
```
