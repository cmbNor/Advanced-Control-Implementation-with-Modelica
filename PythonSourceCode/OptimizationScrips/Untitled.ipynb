```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize

# Define the objective function
def f(x):
    return (x[0] - 1.0) * (x[0] - 1.0) + (x[1] - 2.0) * (x[1] - 2.0)

# New initial guess
initial_guess = [0.0, 0.0]  # You can change these values

# Define the constraints: both x and y should be within the range [-5, 5]
constraint_obj = ({'type': 'ineq', 'fun': lambda x: x[0] - (-5.0)},  # x >= -5
                  {'type': 'ineq', 'fun': lambda x: 5.0 - x[0]},      # x <= 5
                  {'type': 'ineq', 'fun': lambda x: x[1] - (-5.0)},  # y >= -5
                  {'type': 'ineq', 'fun': lambda x: 5.0 - x[1]})      # y <= 5

# Set optimization options (tolerance and maximum iterations)
options = {'tol': 1e-6, 'maxiter': 100}

# Use SciPy's minimize function to find the minimum of the function with constraints and specified options
result = minimize(f, initial_guess, constraints=constraint_obj, options=options)

# Extract the minimum value and corresponding argument
minimum_value = result.fun
minimum_arg = result.x

# Create a range of x and y values for plotting
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

# Create the 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Plot the minimum point on the surface plot
ax.scatter(minimum_arg[0], minimum_arg[1], minimum_value, color='red', marker='o', s=100, label=f'Minimum: ({minimum_arg[0]:.2f}, {minimum_arg[1]:.2f}), Value: {minimum_value:.2f}')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.set_title('3D Surface Plot of f(x) = (x[0] - 1.0)^2 + (x[1] - 2.0)^2')

# Show the plot with the updated legend (without the color bar)
plt.legend()
plt.show()

# Print the new minimum point and value
print(f'New Minimum: ({minimum_arg[0]:.2f}, {minimum_arg[1]:.2f}), New Minimum Value: {minimum_value:.2f}')
```
