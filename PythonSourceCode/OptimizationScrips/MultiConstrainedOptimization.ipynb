```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the cost function
def cost_function(x):
    x0 = x[0]
    x1 = x[1]
    return (x0 - 2) ** 2 + (x1 - 3) ** 2 + 3
```

```python
# Initial guess
initial_guess = np.array([0.0, 0.0])

# Define bounds for each variable
bounds = [(-5, 5), (-5, 5)]  # Bounds for x0 and x1

# Optimize the cost function with bounds using SLSQP
result = minimize(cost_function, initial_guess, method='SLSQP', bounds=bounds)

# Extract the optimized values
optimal_x0 = result.x[0]
optimal_x1 = result.x[1]
min_cost = result.fun
```

```python
# Create a 3D plot of the cost function
x0_range = np.linspace(-5, 5, 100)
x1_range = np.linspace(-5, 5, 100)
X0, X1 = np.meshgrid(x0_range, x1_range)
Z = cost_function([X0, X1])

# Increase the figure size and make the minimum point color stronger
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the cost surface
ax.plot_surface(X0, X1, Z, cmap='viridis', alpha=0.8)

# Plot the optimal point on the 3D surface with a stronger color (red)
ax.scatter(optimal_x0, optimal_x1, min_cost, color='red', s=100, label=f'Optimal Solution ({optimal_x0:.2f}, {optimal_x1:.2f}, {min_cost:.2f})')

# Add a legend
ax.legend()

# Set labels
ax.set_xlabel('X0')
ax.set_ylabel('X1')
ax.set_zlabel('Cost', labelpad=-30)

# Set the title
ax.set_title('Cost Function: (x0 - 2)^2 + (x1 - 3)^2 + 3')

# Save a copy of the plot
plt.savefig('plots/multivariatePythonOptimization.png', dpi=300, bbox_inches='tight')

# Show the 3D plot
plt.show()

# Print the optimized values
print("Optimal x[0] value:", optimal_x0)
print("Optimal x[1] value:", optimal_x1)
print("Minimum cost:", min_cost)
```
