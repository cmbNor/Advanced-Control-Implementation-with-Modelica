```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
# Define the objective function
def f(x):
    return (x - 3.0)**2  # New function (x-3.0)^2
```

```python
# Initial guess
initial_guess = 0.0

# Define the constraints: x should be within the range [-5, 5]
constraint_obj = ({'type': 'ineq', 'fun': lambda x: x - (-5.0)},  # x >= -5
                  {'type': 'ineq', 'fun': lambda x: 5.0 - x})      # x <= 5

# Use SciPy's minimize function to find the minimum of the function with constraints
result = minimize(f, initial_guess, constraints=constraint_obj)

# Extract the minimum value and corresponding argument
minimum_value = result.fun
minimum_arg = result.x[0]  # Extract the scalar value from the numpy array

# Create a range of x values for plotting within the constraint range
x_values = np.linspace(-5.0, 5.0, 100)

# Calculate the corresponding y values using the objective function
y_values = [f(x) for x in x_values]
```

```python
# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Objective Function')
plt.xlabel('x')
plt.ylabel('f(x)')

# Mark the minimum with a red dot
plt.scatter(minimum_arg, minimum_value, color='red', marker='o', zorder=5)

# Add a label for the minimum
plt.annotate(f'Minimum: ({minimum_arg:.2f}, {minimum_value:.2f})', (minimum_arg, minimum_value), textcoords="offset points", xytext=(0,10), ha='center')

# Add the function and constraints to the title
plt.title(f'Constrained Optimization of f(x) = (x - 3.0)^2\nwith Constraints: -5 <= x <= 5')

# Add a legend with the optimized value
plt.legend([f'Objective Function', f'Minimum: ({minimum_arg:.2f}, {minimum_value:.2f})'])
plt.grid(True)

# Save a copy of the plot
plt.savefig('plots/univariatePythonOptimization.png', dpi=300, bbox_inches='tight')

plt.show()
```
